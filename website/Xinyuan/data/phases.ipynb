```python
#count each year's phase crash num and percentage
```

```python
data = np.zeros((115, 10))
phase_df = pd.DataFrame(data, index = range(1900,2015), columns= ['Year ', 'Standing ', 'Pushback / towing ', 'Taxi ', 'Takeoff ', 'Initial climb ', 'En route ', 'Maneuvering ', 'Approach ', 'Landing '])


```

```python
for year in range(1900,2015):
    for phase in phase_df.columns.values:
        #print phase
        sec = new_df[new_df['year']==year]
        #print sec[sec['phase']==phase].count()[0]
        phase_df.loc[year][phase]=sec[sec['phase']==phase].count()[0]
```

```python
phase_df['Total '] = pd.Series(phase_df.sum(axis=1), index=phase_df.index)
phase_df
```

```python
phase_df['Year '] = range(1900,2015)
```

```python
phase_df.to_csv('phase_summary.csv', index=False)
```

```python
phase_percent_df = pd.DataFrame(np.zeros((115, 10)), index = range(1900,2015), columns= ['Year ', 'Standing ', 'Pushback / towing ', 'Taxi ', 'Takeoff ', 'Initial climb ', 'En route ', 'Maneuvering ', 'Approach ', 'Landing '])
print phase_percent_df
```

```python
for year in range(1900,2015):
    for phase in ['Standing ', 'Pushback / towing ', 'Taxi ', 'Takeoff ', 'Initial climb ', 'En route ', 'Maneuvering ', 'Approach ', 'Landing ']:
            percent= phase_df.loc[year][phase]/phase_df.loc[year]['Total ']
            phase_percent_df.loc[year][phase] = percent
```

```python
phase_percent_df['Year '] = range(1900,2015)
phase_percent_df.fillna(0, inplace=True)

```
